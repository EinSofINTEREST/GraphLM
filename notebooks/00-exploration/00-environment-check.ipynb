커널 선택 시 **GraphLM (uv .venv)** 가 표시되어야 합니다.

```python
import sys
from pathlib import Path

import numpy as np

import graphlm

print("python   :", sys.version.split()[0])
print("prefix   :", sys.prefix)
print("graphlm  :", graphlm.__version__)
print("numpy    :", np.__version__)
print("cwd      :", Path.cwd())
```

위 출력의 `prefix` 가 프로젝트 루트의 `.venv` 경로를 가리키면 정상입니다.
